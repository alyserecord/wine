```python
similarity_matrix = cosine_similarity(df)
```

```python
def get_recommendation(wine_name,num_rec=5):
    wine_index = names[names == wine_name].index[0]
    similar_indices = similarity_matrix[wine_index].argsort()[-2:-num_rec-2:-1]
    items = []
    for i in similar_indices:
        items.append(names[i])
    return items
```

```python
wine = 'Veuve Clicquot Yellow Label Brut'
recs = get_recommendation(wine,5)
```

```python
def show_recomendations(wine,wine_recomendations,kmean_label):
    fig,ax = plt.subplots(2,3,figsize=(14,8))
    wine_recomendations.insert(0,wine)
    for i,ax in enumerate(ax.flatten()):
        image = Image.open('../images/{}.jpg'.format(wine_recomendations[i]))
        if i == 0:
            ax.set_xlabel('Selected Wine:\n{}'.format(wine_recomendations[i]),wrap=True,fontsize=16)
        else:
            ax.set_xlabel('Recommendation {}:\n{}'.format(i,wine_recomendations[i]),wrap=True,fontsize=16)            
        ax.set_xticks([]) 
        ax.set_yticks([]) 
        ax.grid()
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.imshow(image)
    plt.tight_layout()
    plt.show()
    return wine_recomendations
```
